<font size='4'>

### Common mistakes

<font size='4'>
    
- Lists are not arrays, so you cannot perform mathematical operations using lists.
    - You can create an array from a list using `np.array()`.
    - You can convert an array to list using `.tolist()`.
- You can only concatenate lists using the `+` operator.


```python
list_a = [1, 2, 3]
print(list_a * 4)

list_b = [4, 5, 6]
print(list_a + list_b)

vec_a = np.array(list_a)
print(vec_a * 4)

list_a_from_vec = vec_a.tolist()
print(list_a_from_vec)
print(type(list_a_from_vec))
print(isinstance(list_a_from_vec, list))
```

## 3. For loop

<font size='4'>
    
- A for loop is a way to iterate over a sequence of elements
- It is a useful tool to automate repetitive tasks
- The syntax is as follows:
    
`for element in sequence:`<br>
`    do something`

```python
var_ls = [1, 2, 3, 4]
for i in var_ls:
    print(i)
```

<font size='4'>

- You can iterate a **list** of elements
    - Numbers, strings, or any other type of object
- You can iterate over a range of numbers, i.e., arrays.
- You can iterate over a list of lists (nested lists).
- You can have for loop embedded in another for loop.

```python
icd_10_cm_codes_ls = [
    ['alchol related disorders', 'opioid related disorders', 'cannabis related disorders'], # 'F10-F19', 
    ['schizophrenia', 'schizotypal disorder', 'delutional disorders', 'brief psychotic disorder'], # 'F20-F29', 
    ['manic episode', 'bipolar disorder', 'depressive episode', 'major depressive disorder, recurrent'] # 'F30-F39', 
]

for code_iter_ls in icd_10_cm_codes_ls:
    print(code_iter_ls)
```

```python
for code_iter_ls in icd_10_cm_codes_ls:
    for disease_iter in code_iter_ls:
        print(disease_iter)
```

### Three ways to iterate in the for loop:

<font size='4'>

- Iterate each element directly.
- Use `range(len(list_name))` with the index of each element.
- Use `enumerate()` function to show both relative index and its corresponding element.

```python
# way 1:
for disease_type_iter in icd_10_code_F20_29_ls:
    print('The F20-29 codes include {} disease type.'.format(disease_type_iter))
```

```python
# way 2:
for num_iter in range(len(icd_10_code_F20_29_ls)):
    print('The F20-29 codes include {} disease type.'.format(icd_10_code_F20_29_ls[num_iter]))  
```

```python
for num_iter in range(len(icd_10_code_F20_29_ls)):
    print('The F20-29 codes include {} disease type in the location {}.'.format(
        icd_10_code_F20_29_ls[num_iter], num_iter+1)
    )
```

```python
# way 3:
# we could also use enumerate() to show both relative index and its corresponding element.
for id_iter, disease_iter in enumerate(icd_10_code_F20_29_ls):
    print('The F20-29 codes include {} disease type in the location {}.'.format(
        disease_iter, id_iter+1)
    )
```

## 4. While loop

<font size='4'>

- A while loop is to execute a block of codes repeatedly until a given condition is satisfied.
- When the condition is `False`, the codes within the while loop is no longer executed, <br> instead, the line immediately after the loop is executed.
- General syntax:

`initialization` <br>
`related to expression` <br>
`while expression:`<br>
`    statements`

```python
# simple example
i = 1
while i < 6:
    print(i)
    i += 1
```

```python
# Avoid writing this while loop in python.
'''
age = 28

# the test condition is always True
while age > 19:
    print('Infinite Loop')
# How do you stop it in Jupyter notebook?
# Press the "Interrupt the kernel" button.
'''
# You can use ''' ''' for block comment.
```

### Advanced features with while loop

<font size='4'>

- `break`: immediately terminate a loop.
- `continue`: end withthe current iteration, can be used as to skip tasks for certain iterations.
- `end`: optional clause, only run when the loop terminates naturally.

```python
# while with break
number_iter = 6
while number_iter > 0:
    number_iter -= 1
    if number_iter == 2:
        break
    print(number_iter)
print('loop ended')
```

```python
# while with break
# What if we move the print function early?
number_iter = 6
while number_iter > 0:
    number_iter -= 1
    print(number_iter)
    if number_iter == 2:
        break
print('loop ended')
```

```python
# while with continue, 
# skip tasks within an iteration
number_iter = 6

while number_iter > 0:
    number_iter -= 1
    if number_iter in [1,2]:
        continue
    print(number_iter)

print("Loop ended")
```

```python
# while with else
# only run else in presence of natural termination
i = 0
while i < 4:
    i += 1
    print(i)
else:  # Executed because no break in for
    print("No Break")
```

```python
i = 0
while i < 4:
    i += 1
    print(i)
    break
else:  # Not executed as there is a break
    print("No Break")
```
